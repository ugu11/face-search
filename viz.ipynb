```python
import torch
from torchvision.io import read_image
from torchvision.transforms.functional import to_pil_image, crop
from einops.layers.torch import Rearrange
```

```python
img = read_image("ultron-face.jpg")
```

```python
img.shape
```

```python
img = crop(img, 10, 0, 196, 196)
to_pil_image(img)
```

```python
(img / 255) * 2 - 1
```

```python
patchify = Rearrange('b c (h p1) (w p2) -> b (h w) (p1 p2 c)', p1 = 49, p2 = 49)
patches = patchify(img.unsqueeze(0))
patches.shape
```
